```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with movie data
data = {
    'title': ['Christmas Joy', 'Holiday Romance', 'The Christmas Chronicles', 'Miracle on 34th Street'],
    'country': ['USA', 'USA', 'USA', 'USA'],
    'date': ['2023-12-10', '2022-12-15', '2021-11-22', '2019-12-14']
}

# Create a DataFrame from the sample data
release_dates = pd.DataFrame(data)
release_dates['date'] = pd.to_datetime(release_dates['date'])

# Filter for movies with "Christmas" in their title released in the USA
christmas_movies_usa = release_dates[(release_dates['title'].str.contains('Christmas')) & (release_dates['country'] == 'USA')]

# Extract the month from the release date
christmas_movies_usa['release_month'] = christmas_movies_usa['date'].dt.month_name()

# Count the number of movies released in each month
month_counts = christmas_movies_usa['release_month'].value_counts().reindex([
    'January', 'February', 'March', 'April', 'May', 'June', 'July', 'August',
    'September', 'October', 'November', 'December'
])

# Create a bar plot
plt.figure(figsize=(10, 6))
month_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.title('Movies with "Christmas" in Title Released in the USA by Month')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# Sample DataFrame with movie data
data = {
    'title': ['Romantic Comedy', 'Summer Romance', 'Drama and Romance', 'Love Story', 'Holiday Romance'],
    'country': ['USA', 'USA', 'USA', 'USA', 'USA'],
    'date': ['2023-03-15', '2023-05-20', '2023-08-10', '2023-02-14', '2023-11-28']
}

# Create a DataFrame from the sample data
release_dates = pd.DataFrame(data)
release_dates['date'] = pd.to_datetime(release_dates['date'])
```

```python
import matplotlib.pyplot as plt

# Filter for movies with "Romance" in their title released in the USA
romance_movies_usa = release_dates[(release_dates['title'].str.contains('Romance')) & (release_dates['country'] == 'USA')]

# Extract the day of the week from the release date
romance_movies_usa['release_day_of_week'] = romance_movies_usa['date'].dt.day_name()

# Count the number of movies released on each day of the week
day_of_week_counts = romance_movies_usa['release_day_of_week'].value_counts().reindex(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Create a bar plot
plt.figure(figsize=(10, 6))
day_of_week_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Movies')
plt.title('Sample: Movies with "Romance" in Title Released in the USA by Day of the Week')
plt.xticks(rotation=45)
plt.show()
```
