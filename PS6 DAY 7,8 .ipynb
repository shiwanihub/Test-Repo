```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    # Function to add a new element to the end of the linked list
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
        else:
            current = self.head
            while current.next:
                current = current.next
            current.next = new_node

    # Function to find the smallest element in the linked list
    def find_smallest(self):
        if not self.head:
            return None

        current = self.head
        smallest = current.data

        while current:
            if current.data < smallest:
                smallest = current.data
            current = current.next

        return smallest

    # Function to insert an element if it's not a duplicate
    def insert_unique(self, data):
        if not self.head:
            self.head = Node(data)
            return

        current = self.head
        while current.next:
            if current.data == data:
                return  
            current = current.next

        if current.data == data:
            return 
        current.next = Node(data)

    
    def display_reverse(self):
        if not self.head:
            return []

        stack = []
        current = self.head
        while current:
            stack.append(current.data)
            current = current.next

        reversed_elements = []
        while stack:
            reversed_elements.append(stack.pop())

        return reversed_elements

# Create a linked list and perform operations
if __name__ == "__main__":
    linked_list = LinkedList()

    # Create the list
    elements = [10, 5, 15, 7, 5, 22, 7]
    for element in elements:
        linked_list.append(element)

    # Find the smallest element
    smallest = linked_list.find_smallest()
    print("Smallest element:", smallest)

    # Insert a unique element
    linked_list.insert_unique(3)
    
    # Display elements in reverse order
    reversed_elements = linked_list.display_reverse()
    print("Elements in reverse order:", reversed_elements)
```
