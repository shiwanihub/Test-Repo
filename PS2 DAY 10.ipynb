```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    "year": [
        1920, 1921, 1931, 1932, 1933, 1940, 1941, 1942, 1943, 1950,
        1951, 1952, 1959, 1960, 1961, 1968, 1969, 1970, 1971, 1980,
        1981, 1982, 1989, 1990, 1991, 1999, 2000, 2001, 2009, 2010, 2011, 2012, 2013 , 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023
    ]
}

df = pd.DataFrame(data)

# Calculate the decade for each movie release year
df['decade'] = (df['year'] // 10) * 10

# Count the number of films released each decade
decade_counts = df['decade'].value_counts().sort_index()

# Plot the number of films released each decade
plt.figure(figsize=(10, 6))
plt.bar(decade_counts.index.astype(str), decade_counts.values)
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd


data = {
    "name": [
        "Actor 1",
        "Actress 1",
        "Actor 2",
        "Actress 2",
        "Actor 3",
        "Actress 3",
        "Actor 4",
        "Actress 4",
        "Actor 5",
        "Actress 5"
    ],
    "Character": [
        "Role 1",
        "Role 2",
        "Role 3",
        "Role 4",
        "Role 5",
        "Role 6",
        "Role 7",
        "Role 8",
        "Role 9",
        "Role 10"
    ],
    "n": [
        1,
        1,
        2,
        1,
        1,
        2,
        1,
        1,
        1,
        2
    ],
    "year": [
        1950,
        1950,
        1951,
        1951,
        1952,
        1952,
        1953,
        1953,
        1954,
        1954
    ]
}

df = pd.DataFrame(data)

# Filter for leading roles (n=1) for actors and actresses separately
leading_roles_actors = df[(df['n'] == 1) & (df['name'].str.contains('Actor'))]
leading_roles_actresses = df[(df['n'] == 1) & (df['name'].str.contains('Actress'))]

# Count the number of leading roles for actors and actresses in the 1950s
count_leading_roles_actors = len(leading_roles_actors[leading_roles_actors['year'] >= 1950])
count_leading_roles_actresses = len(leading_roles_actresses[leading_roles_actresses['year'] >= 1950])

print(f"Number of leading (n=1) roles for actors in the 1950s: {count_leading_roles_actors}")
print(f"Number of leading (n=1) roles for actresses in the 1950s: {count_leading_roles_actresses}")
```
